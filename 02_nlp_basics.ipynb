### What is Natural Language Processing

**Natural Language Processing (NLP)** is a field within artificial intelligence that focuses on enabling computers to understand, interpret, and generate human language in a way that feels natural to people. At its core, NLP involves working with **raw text data understanding**, which means taking unstructured, original language data (like sentences, paragraphs, or documents) and transforming it into a format computers can analyze and process.

#### Why is NLP Important?

1. **Handling Raw Text Data**: Much of the worlds data exists as raw, unstructured textsuch as emails, customer reviews, social media posts, or support tickets. NLP makes it possible to automatically understand and extract meaningful information from this raw text data, which otherwise would be overwhelming to process manually.

2. **Improving Human-Computer Interaction**: NLP powers voice assistants, chatbots, and other interactive technologies that allow people to communicate naturally with machines using everyday language, rather than rigid commands.

3. **Automating Language-Based Tasks**: NLP enables automation of text-intensive tasks like sorting emails, categorizing customer inquiries, and extracting key information from documents, reducing manual effort and improving efficiency.

4. **Enabling Advanced Data Analysis**: By processing raw text data, NLP allows organizations to extract insights from vast amounts of unstructured language data, making it easier to understand customer sentiment, trends, and feedback, and supporting better decision-making.

In short, NLP is essential for transforming raw text data into structured, usable insights, allowing computers to bridge the gap between human language and digital data processing. This foundational step of raw text understanding makes possible a range of applications, from automated customer support to sophisticated data analytics.

 ### `token.pos`:
   - This is the **universal part-of-speech tag** (a number or an integer) assigned to the token. It corresponds to the token's part-of-speech in the **universal POS tagging scheme**, which is a simplified POS tag set.
   - Example: `token.pos` might return `96` for a noun, `87` for a verb, etc.

````python
for token in doc:
    print(f'{token.text:20} {token.pos:10}')     #Indentation Applied to using f-string
````

### `token.pos_`:
   - This is the **string representation** of the universal part-of-speech tag.
   - Example: `token.pos_` might return `"NOUN"` for a noun, `"VERB"` for a verb, etc.

````python
for token in doc:
    print(f'{token.text:20} {token.pos:20} {token.pos_:20}')      #Indentation Applied to using f-string
````

### Explanation:
- **`token.dep_`**: This is the **string representation** of the syntactic dependency relation between the token and its head.
  - Examples: `"nsubj"` (nominal subject), `"dobj"` (direct object), `"amod"` (adjectival modifier), `"xcomp"` (open clausal complement).
  
- **`token.dep`**: This is the **integer code** that corresponds to the syntactic dependency relation.
  - These integers map directly to the string representation in `token.dep_`.

### Additional Dependency Tags and their Meanings:
- **`vocative`**: A noun or phrase that addresses someone directly (e.g., "Hello, John!").
- **`mark`**: A word introducing a subclause (e.g., "if", "because").
- **`dislocated`**: A part of the sentence that has been moved, often for emphasis (e.g., "As for the cake, it's delicious").
- **`parataxis`**: Two clauses or phrases that are loosely connected, often by punctuation like commas or semicolons.
- **`ccomp`**: A clausal complement, often used to indicate verb complements (e.g., "She said [that she would go]").
- **`corrdet`**: Correlative determiners or conjunctions used in pairs (e.g., "either...or", "neither...nor").

````python
import spacy
import pandas as pd
from IPython.display import display, HTML

# Load the spaCy model (replace with your specific model)
nlp = spacy.load("en_core_web_sm")

# Process your text
# text = "Apple is considering acquiring a European company for 3 billion next year."  # Example sentence
# doc = nlp(text)
doc = nlp(u"Apple is considering acquiring a European company for 3 billion next year.") # u Refers to Unicode String

# Create a list of dictionaries with token attributes
data = []
for token in doc:
    data.append({
        # Most commonly used attributes (left)
        'Token': token.text,                    # The token's original form
        'Lemma': token.lemma_,                   # The base form of the token
        'POS': token.pos_,                       # Part of speech (coarse)
        'Tag': token.tag_,                       # Part of speech (fine-grained)
        'Dep': token.dep_,                       # Dependency relation to the head token
        'Shape': token.shape_,                   # The token's shape (capitalization, digits, etc.)
        'Is Alpha': token.is_alpha,              # Whether the token is an alphabetic string
        'Is Stop': token.is_stop,                # Whether the token is a stop word
        'Is Punct': token.is_punct,              # Whether the token is punctuation
        'Is Digit': token.is_digit,              # Whether the token is a number
        'Is Space': token.is_space,              # Whether the token is a space
        'Ent Type': token.ent_type_,             # Named entity type (if any)
        'Vector (first 5)': str(token.vector[:5]), # The token's vector representation (first 5 values)

        # Less commonly used attributes (right)
        'Is OOV': token.is_oov,                  # Whether the token is out-of-vocabulary
        'Lang': token.lang_,                     # The language of the token
        'Sent Start': token.sent_start,           # Whether the token is the start of the sentence
        'Prefix': token.prefix_,                 # Prefix of the token
        'Suffix': token.suffix_,                 # Suffix of the token
        'Norm': token.norm_,                     # Normalized form of the token
        'Left Edge': token.left_edge,             # Leftmost token in the subtree of the token
        'Right Edge': token.right_edge,           # Rightmost token in the subtree of the token
        'Subtree': list(token.subtree),          # Tokens in the syntactic subtree of the token
        'Ancestors': list(token.ancestors),      # Tokens in the syntactic ancestors of the token
    })

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Rearrange columns so that the most common attributes are on the left
columns_order = [
    'Token', 'Lemma', 'POS', 'Tag', 'Dep', 'Shape', 'Is Alpha', 'Is Stop', 'Is Punct', 
    'Is Digit', 'Is Space', 'Ent Type', 'Vector (first 5)', 
    'Is OOV', 'Lang', 'Sent Start', 'Prefix', 'Suffix', 'Norm', 
    'Left Edge', 'Right Edge', 'Subtree', 'Ancestors'
]

df = df[columns_order]

# Display the DataFrame in a scrollable format
display(HTML(df.to_html(index=False)))

# Add custom CSS for making the table scrollable and displaying it in a block
display(HTML('<style>.dataframe {display: block; overflow-x: auto; max-width: 100%;}</style>'))
````
